Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation
of 5 using Python. Interpret the results.

```python
import scipy.stats as stats

# Given data
mean = 50
std_dev = 5
sample_size = 30

# Calculate the standard error
std_error = std_dev / (sample_size ** 0.5)

# Calculate the confidence interval
confidence_interval = stats.norm.interval(0.95, loc=mean, scale=std_error)
confidence_interval
```

Q2. Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag
matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20%
brown. Use Python to perform the test with a significance level of 0.05.

```python
import numpy as np

# Observed counts
observed_counts = np.array([23, 18, 19, 8, 12, 20])

# Total number of M&Ms
total_count = np.sum(observed_counts)

# Expected proportions
expected_proportions = np.array([0.20, 0.20, 0.20, 0.10, 0.10, 0.20])

# Expected counts
expected_counts = total_count * expected_proportions

# Chi-square test
chi_square_stat, p_value = stats.chisquare(observed_counts, f_exp=expected_counts)

chi_square_stat, p_value
```

Q4. A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
import math

# Given data
x = 60  # number of individuals who smoke
n = 500  # total number of individuals

# Sample proportion
p_hat = x / n

# Standard error
se = math.sqrt(p_hat * (1 - p_hat) / n)

# Z-score for 95% confidence
z = 1.96

# Confidence interval
ci_lower = p_hat - z * se
ci_upper = p_hat + z * se

(ci_lower, ci_upper)
```

Q5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

```python
mean = 75
std_dev = 12
sample_size = 30

# Calculate the standard error
std_error = std_dev / math.sqrt(sample_size)

# Z-score for 90% confidence
z = 1.645

# Confidence interval
ci_lower = mean - z * std_error
ci_upper = mean + z * std_error

(ci_lower, ci_upper)
```

Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

```python
x = 520  # number of individuals who prefer Coke
n = 1000  # total number of individuals

# Sample proportion
p_hat = x / n

# Standard error
se = math.sqrt(p_hat * (1 - p_hat) / n)

# Z-score for 99% confidence
z = 2.576

# Confidence interval
ci_lower = p_hat - z * se
ci_upper = p_hat + z * se

(ci_lower, ci_upper)
```

Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

```python
sample_mean = 72
population_mean = 70
sample_std_dev = 10
sample_size = 30

# Calculate the test statistic
t_statistic = (sample_mean - population_mean) / (sample_std_dev / (sample_size ** 0.5))

# Degrees of freedom
df = sample_size - 1

# Calculate the p-value
p_value = 2 * stats.t.sf(abs(t_statistic), df)

t_statistic, p_value
```
