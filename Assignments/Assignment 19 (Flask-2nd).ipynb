### Q1. Explain GET and POST methods.

ANS: 
1. GET Method: The GET method is used to retrieve or fetch data from a server. When a client makes a GET request, it includes parameters in the URL itself. These parameters are appended to the URL as query strings and are visible to everyone, including in browser history and server logs. GET requests are typically used to retrieve data from a server without modifying it.
2. POST Method: The POST method is used to send data to the server to create, update, or modify a resource. Unlike GET requests, the parameters and data sent with a POST request are not visible in the URL. Instead, the data is sent in the body of the HTTP request. This makes POST requests more suitable for sending sensitive information, such as passwords or credit card details, as the data is not directly visible.

### Q2. Why is request used in Flask?

ANS: In Flask, the request object is used to access and handle incoming HTTP requests made by clients (such as web browsers) to the Flask server. It provides convenient methods and attributes to extract data from the request, access headers, query parameters, form data, cookies, and other relevant information associated with the request.

### Q3. Why is redirect() used in Flask?

ANS: The redirect() function is used to perform a redirection to a different URL or route within the application. It allows you to redirect the client's browser to a different page after processing a request or completing a specific action.

### Q4. What are templates in Flask? Why is the render_template() function used?

ANS: In Flask, templates are files that contain dynamic content and are used to generate HTML pages dynamically. Templates provide a way to separate the presentation logic from the application logic, making it easier to maintain and organize the code.

The render_template() fuction is used for the following reasons:
1. Template File Location
2. Passing the Data
3. Template rendering
