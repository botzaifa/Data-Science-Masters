### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

**ANS:** Seaborn is a powerful visualization library in Python built on top of Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Here are five plots that can be created using Seaborn, along with their uses:

1. **Scatter Plot (`scatterplot`)**:
   - Use: A scatter plot displays values for two variables for a set of data. It helps in identifying the relationship between two continuous variables and in detecting outliers.


2. **Bar Plot (`barplot`)**:
   - Use: A bar plot shows the relationship between a categorical variable and a continuous variable. It is useful for comparing the quantities of different categories.


3. **Box Plot (`boxplot`)**:
   - Use: A box plot (or whisker plot) provides a summary of a set of data values, showing their central tendency and variability. It is used to detect outliers and understand the distribution of the data.


4. **Heatmap (`heatmap`)**:
   - Use: A heatmap displays data in matrix form, using color to indicate the value of each cell. It is useful for visualizing the correlation between variables or the density of data points.


5. **Pair Plot (`pairplot`)**:
   - Use: A pair plot (also known as a scatterplot matrix) displays pairwise relationships in a dataset. It is useful for quickly visualizing the interactions between multiple variables and identifying potential patterns.



### Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

plt.figure(figsize=(12, 6))
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Add title and labels
plt.title("FMRI Signal over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Show the plot
plt.show()
```

### Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

sns.boxplot(ax=axes[0], data=titanic, x='pclass', y='age')
axes[0].set_title('Age Distribution by Passenger Class')
axes[0].set_xlabel('Passenger Class')
axes[0].set_ylabel('Age')

sns.boxplot(ax=axes[1], data=titanic, x='pclass', y='fare')
axes[1].set_title('Fare Distribution by Passenger Class')
axes[1].set_xlabel('Passenger Class')
axes[1].set_ylabel('Fare')

# Show the plot
plt.tight_layout()
plt.show()
```

### Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

plt.figure(figsize=(12, 6))
sns.histplot(data=diamonds, x='price', hue='cut', kde=True, multiple="stack")

plt.title('Price Distribution of Diamonds by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

### Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

sns.pairplot(iris, hue='species', diag_kind='kde')

plt.suptitle('Pair Plot of Iris Dataset by Species', y=1.02)

# Show the plot
plt.show()
```

### Q6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

flights_pivot = flights.pivot("month", "year", "passengers")

plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

plt.title('Heatmap of Monthly Flight Passengers')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
