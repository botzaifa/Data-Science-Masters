### Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

**ANS:** 

1. **Collect Data:**
   - You have data on two variables: study time (in hours) and final exam scores.
   - For each student, you'll have a pair of values: study time and exam score.

2. **Calculate Pearson Correlation Coefficient (r):**
   - The formula for Pearson correlation coefficient is:
     $$ r = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sqrt{\sum{(x_i - \bar{x})^2} \sum{(y_i - \bar{y})^2}}} $$
   - Where:
     - \(x_i\) represents study time for each student.
     - \(y_i\) represents exam scores for each student.
     - \(bar{x}\) is the mean study time.
     - \(bar{y}\) is the mean exam score.


3. **Interpretation:**
   - If \(r\) is close to 1, it indicates a strong positive linear relationship. As study time increases, exam scores tend to increase.
   - If \(r\) is close to -1, it indicates a strong negative linear relationship. As study time increases, exam scores tend to decrease.
   - If \(r\) is close to 0, there's little or no linear relationship.
   - Remember, correlation doesn't imply causation! Other factors could be at play.

4. **Example Interpretation:**
   - If \(r = 0.7\), you can say, "There is a strong positive correlation (0.7) between study time and exam scores. Students who study more tend to perform better on exams."


### Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two variables and interpret the result.

**ANS:** Calculating Spearman's Rank Correlation Coefficient:

1. **Rank the data:** Assign ranks to the sleep hours and job satisfaction values. The highest value gets rank 1, the second highest rank 2, and so on. If there are ties, assign the average rank to the tied values.
2. **Calculate the difference in ranks (d) for each pair of observations.**
3. **Square the differences (d^2).**
4. **Sum the squared differences (Σd^2).**
5. **Calculate the Spearman's rank correlation coefficient (rs) using the formula:**

   ```
   rs = 1 - (6 * Σd^2) / (n * (n^2 - 1))
   ```
   where:
   * `rs` is the Spearman's rank correlation coefficient
   * `Σd^2` is the sum of the squared differences in ranks
   * `n` is the number of observations

**`Interpreting the Results:`**

* **rs = 1:** Perfect positive monotonic relationship (as one variable increases, the other consistently increases)
* **rs = -1:** Perfect negative monotonic relationship (as one variable increases, the other consistently decreases)
* **rs = 0:** No monotonic relationship between the variables




### Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.

**ANS:** 

1. **`Pearson Correlation Coefficient (r)`:**
   - The Pearson correlation coefficient measures the linear relationship between two continuous variables.
   - It ranges from -1 (perfect negative correlation) to 1 (perfect positive correlation).
   - If the value is close to 0, there's little or no linear relationship.
   - To calculate it:
     $$ r = \frac{\sum{(x_i - \bar{x})(y_i - \bar{y})}}{\sqrt{\sum{(x_i - \bar{x})^2} \sum{(y_i - \bar{y})^2}}} $$
   - Where:
     - \(x_i\) and \(y_i\) are the data points for exercise hours and BMI, respectively.
     - \(bar{x}\) and \(bar{y}\) are the means of exercise hours and BMI.



2. **`Spearman's Rank Correlation (ρ)`:**
   - Spearman's rank correlation assesses the monotonic relationship (not necessarily linear) between two variables.
   - It's based on ranks rather than actual values.
   - Ranges from -1 (perfect negative monotonicity) to 1 (perfect positive monotonicity).
   - To calculate it:
     $$ \rho = 1 - \frac{6\sum{d_i^2}}{n(n^2 - 1)} $$
   - Where:
     - \(d_i\) is the difference in ranks between exercise hours and BMI for each participant.
     - \(n\) is the sample size (50 in your case).


3. **`Comparison:`**
   - If both correlations are significant:
     - If \(|r| > |rho|\), there's a stronger linear relationship.
     - If \(|rho| > |r|\), there's a stronger monotonic relationship.
   - If one is significant and the other isn't:
     - Consider the significant one as more reliable.

### Q6. A company is interested in examining the relationship between the number of sales calls made per day and the number of sales made per week. The company collected data on both variables from a sample of 30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.

**ANS:**  **`Steps Involved in Calculating the Pearson Correlation Coefficient:`**
1. **Collect the data:** Gather the necessary information from the 30 sales representatives.
2. **Calculate the mean and standard deviation:** Determine these values for both the number of sales calls and the number of sales.
3. **Calculate the covariance:** Measure the joint variability of the two variables.
4. **Calculate the Pearson correlation coefficient:** Use the formula:

  $$ r = covariance(X, Y) / (standard deviation(X) * standard deviation(Y)) $$

  where:
  * r is the Pearson correlation coefficient
  * X is the number of sales calls
  * Y is the number of sales

**`Interpreting the Results:`**
- **r ranges from -1 to 1:**
  * A value close to 1 indicates a strong positive correlation (as one variable increases, the other tends to increase).
  * A value close to -1 indicates a strong negative correlation (as one variable increases, the other tends to decrease).
  * A value close to 0 indicates a weak or no correlation between the variables.
