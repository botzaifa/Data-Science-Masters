# Data Interpolation
Process of estimatimating unknown values within a dataset based on known values.

## 1. Linear Interpolation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d #'1D' means interpolation in one dimension
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
plt.scatter(x,y)
```

```python
# Interpolation of the data
## syntax ->  x=np.linspace(Range_Start,Range_end,no of values to be added) 
x_new = np.linspace(1,5,10) # creating new x values
y_new = np.interp(x_new,x,y)
```

```python
y_new
```

```python
plt.scatter(x_new,y_new)
```

## 2. Cubic Interpolation using Scipy
in Cubic Interpolation the values of one of the variables is the cube of the other

```python
a = np.array([1,8,27,64,125])
```

```python
plt.scatter(x,a)
```

```python
# create a cubic interpolation
f = interp1d(x,a,kind='cubic')
```

```python
# Interpolation of the data
a_new = f(x_new)
```

```python
a_new
```

```python
plt.scatter(x_new,a_new)
```

## 3. Polynomial Interpolation

```python
# Taking the data from the linear interpolation section
plt.scatter(x,y)
```

```python
# Interpolation of the data
p = np.polyfit(x,y,2)
```

```python
y_polyint = np.polyval(p, x_new)
```

```python
plt.scatter(x_new,y_polyint)
```
