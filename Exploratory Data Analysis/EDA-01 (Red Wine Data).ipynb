```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("E:\Datasets\Exploratory analysis\winequality-red.csv")
df.head()
```

```python
# Summary of the dataset (info)
df.info()
```

```python
df['quality'].unique()
```

```python
# Check for missing values in the dataset
df.isnull().sum()
```

```python
# Check for duplicate records
df[df.duplicated()]
```

```python
# Deleting the duplicated values
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
# Finding the correlation
df.corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
```

```python
#Checking if the data is imbalanced or not
df.quality.value_counts()
```

```python
df.quality.value_counts().plot(kind='bar')
plt.xlabel("wine quality")
plt.ylabel('count')
plt.show()

# Now we know that it is an imbalanced dataset
```

```python
for column in df.columns:
    sns.histplot(df[column],kde=True)
```

```python
sns.histplot(df['alcohol'])
```

```python
#univariate,bivariate,multivariate analysis
sns.pairplot(df)
```

```python
# Categorical Plot
sns.catplot(x='quality',y='alcohol',data=df,kind='box')
```

```python
sns.scatterplot(x='alcohol',y='pH',hue='quality',data=df)
```
